```python


from bs4 import BeautifulSoup
import re

# Read the HOCR file
with open("/content/OCR-D-HOCR_OCR-D-HOCR_00001.xml", "r", encoding="utf-8") as file:
    hocr_content = file.read()

# Parse the HOCR content
soup = BeautifulSoup(hocr_content, "html.parser")

div_title = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 286 108 869 249" in value)

# Extract text from the first div
if div_title:
    title = div_title.get_text(strip=True)
else:
    title = ""
# Find the div with class "ocr_carea" and title attribute containing "bbox 143 1454 1006 1529"
div_place_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 143 1454 1006 1529" in value)

# Extract text from the third div with the specified format (\b[A-Z][a-z]+\b)
if div_place_tag:
    place_text = ""
    text = div_place_tag.get_text(strip=True)
    match = re.search(r'\bLeipzig\b', text)
    if match:
        place_text = match.group(0)
else:
    place_text = ""
# Find the div with class "ocr_carea" and title attribute containing "bbox 143 1454 1006 1529"
div_org_name_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 143 1454 1006 1529" in value)

# Extract text from the third div with the specified format for orgName
if div_org_name_tag:
    org_name_text = ""
    text = div_org_name_tag.get_text(strip=True)
    match = re.search(r"B.G. Teubner", text)
    if match:
        org_name_text = match.group(0)
else:
    org_name_text = ""

div_p1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 146 252 1009 312" in value)

# Extract text from the first div
if div_p1:
    p1 = div_p1.get_text(strip=True)
else:
    p1 = ""

div_p2 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 142 371 1010 526" in value)

# Extract text from the first div
if div_p2:
    p2 = div_p2.get_text(strip=True)
else:
    p2 = ""

div_p3 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 146 530 897 657" in value)

# Extract text from the first div
if div_p3:
    p3 = div_p3.get_text(strip=True)
else:
    p3 = ""

div_p4 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 142 700 1009 888" in value)

# Extract text from the first div
if div_p4:
    p4 = div_p4.get_text(strip=True)
else:
    p4 = ""

div_h1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 145 907 301 944" in value)

# Extract text from the first div
if div_h1:
    h1 = div_h1.get_text(strip=True)
else:
    h1 = ""


div_p5 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 145 954 1009 1018" in value)

# Extract text from the first div
if div_p5:
    p5 = div_p5.get_text(strip=True)
else:
    p5 = ""

div_p6 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 141 1027 1008 1084" in value)

# Extract text from the first div
if div_p6:
    p6 = div_p6.get_text(strip=True)
else:
    p6 = ""

div_p7 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 140 1095 1007 1264" in value)

# Extract text from the first div
if div_p7:
    p7 = div_p7.get_text(strip=True)
else:
    p7 = ""

div_p8 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 140 1273 1007 1441" in value)

# Extract text from the first div
if div_p8:
    p8 = div_p8.get_text(strip=True)
else:
    p8 = ""


# Create TEI XML document
tei_xml = f"""
<TEI>
  <teiHeader>
    <fileDesc>
    <titleStmt>
        <title>{title}</title>
        </titleStmt>
        <orgName>{org_name_text}</orgName>
      <publicationStmt>
        <publPlace>{place_text}</publPlace>
      </publicationStmt>
    </fileDesc>
  </teiHeader>
  <text>
    <body>
      <div>
        <p>{p1}</p>
        <p>{p2}</p>
        <p>{p3}</p>
        <p>{p4}</p>
        <head>{h1}</head>
        <p>{p5}</p>
        <p>{p6}</p>
        <p>{p7}</p>
        <p>{p8}</p>

      </div>
    </body>
  </text>
</TEI>
"""

# Write TEI XML document to file
with open("output_141739.tei.xml", "w", encoding="utf-8") as file:
    file.write(tei_xml)
```
