```python


from bs4 import BeautifulSoup
import re

# Read the HOCR file
with open("/content/OCR-D-HOCR_deu_damenschneiderin_1941_bb.xml", "r", encoding="utf-8") as file:
    hocr_content = file.read()

# Parse the HOCR content
soup = BeautifulSoup(hocr_content, "html.parser")

# Define a list of target IDs
target_ids = ["bbox 351 257 1381 304", "bbox 563 302 1172 353", "bbox 367 353 1364 395"]

# Initialize a variable to store the extracted text
title = ""

# Loop over each target ID
for target_id in target_ids:
    # Find the div element with the specified class and title
    div = soup.find("div", class_="ocr_carea", title=lambda value: value and target_id in value)

    # Extract text from the div if found
    if div:
        title += div.get_text(strip=True) + "\n"
# Find the div with class "ocr_carea" and title attribute containing "bbox 222 2337 1531 2374"
div_place_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 222 2337 1531 2374" in value)

# Extract text from the third div with the specified format (\b[A-Z][a-z]+\b)
if div_place_tag:
    place_text = ""
    text = div_place_tag.get_text(strip=True)
    match = re.search(r'Bielefeld', text)
    if match:
        place_text = match.group(0)
else:
    place_text = ""
# Find the div with class "ocr_carea" and title attribute containing "bbox 222 2337 1531 2374"
div_org_name_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 222 2337 1531 2374" in value)

# Extract text from the third div with the specified format for orgName
if div_org_name_tag:
    org_name_text = ""
    text = div_org_name_tag.get_text(strip=True)
    match = re.search(r"KG Bielefeld", text)
    if match:
        org_name_text = match.group(0)
else:
    org_name_text = ""

div_p1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 210 498 509 537" in value)

# Extract text from the first div
if div_p1:
    p1 = div_p1.get_text(strip=True)
else:
    p1 = ""

div_h1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 211 575 454 617" in value)

# Extract text from the first div
if div_h1:
    h1 = div_h1.get_text(strip=True)
else:
    h1 = ""

div_p2 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 418 625 1526 734" in value)

# Extract text from the first div
if div_p2:
    p2 = div_p2.get_text(strip=True)
else:
    p2 = ""

div_h2 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 214 768 1397 811" in value)

# Extract text from the first div
if div_h2:
    h2 = div_h2.get_text(strip=True)
else:
    h2 = ""

div_p3 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 211 835 1527 873" in value)

# Extract text from the first div
if div_p3:
    p3 = div_p3.get_text(strip=True)
else:
    p3 = ""

div_p4 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 420 871 1532 1540" in value)

# Extract text from the first div
if div_p4:
    p4 = div_p4.get_text(strip=True)
else:
    p4 = ""

div_h3 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 217 1574 405 1610" in value)

# Extract text from the first div
if div_h3:
    h3 = div_h3.get_text(strip=True)
else:
    h3 = ""


div_p5 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 426 1572 1170 1643" in value)

# Extract text from the first div
if div_p5:
    p5 = div_p5.get_text(strip=True)
else:
    p5 = ""


# Create TEI XML document
tei_xml = f"""
<TEI>
  <teiHeader>
    <fileDesc>
    <titleStmt>
        <title>{title}</title>
        </titleStmt>
        <orgName>{org_name_text}</orgName>
      <publicationStmt>
        <publPlace>{place_text}</publPlace>
      </publicationStmt>
    </fileDesc>
  </teiHeader>
  <text>
    <body>
      <div>
        <p>{p1}</p>
        <head>{h1}</head>
        <p>{p2}</p>
        <head>{h2}</head>
        <p>{p3}</p>
        <p>{p4}</p>
        <head>{h3}</head>
        <p>{p5}</p>
      </div>
    </body>
  </text>
</TEI>
"""

# Write TEI XML document to file
with open("output_deu_damenschiderin.tei.xml", "w", encoding="utf-8") as file:
    file.write(tei_xml)
```
