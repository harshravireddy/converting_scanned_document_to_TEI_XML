```python


from bs4 import BeautifulSoup
import re

# Read the HOCR file
with open("/content/combined_hocr_141813.hocr", "r", encoding="utf-8") as file:
    hocr_content = file.read()

# Parse the HOCR content
soup = BeautifulSoup(hocr_content, "html.parser")

div_title = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 270 119 854 259" in value)

# Extract text from the first div
if div_title:
    title = div_title.get_text(strip=True)
else:
    title = ""
# Find the div with class "ocr_carea" and title attribute containing "bbox 111 1497 978 1538"
div_place_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 111 1497 978 1538" in value)

# Extract text from the third div with the specified format (\b[A-Z][a-z]+\b)
if div_place_tag:
    place_text = ""
    text = div_place_tag.get_text(strip=True)
    match = re.search(r'\bLeipzig\b', text)
    if match:
        place_text = match.group(0)
else:
    place_text = ""
# Find the div with class "ocr_carea" and title attribute containing "bbox 111 1497 978 1538"
div_org_name_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 111 1497 978 1538" in value)

# Extract text from the third div with the specified format for orgName
if div_org_name_tag:
    org_name_text = ""
    text = div_org_name_tag.get_text(strip=True)
    match = re.search(r"B.G. Teubner", text)
    if match:
        org_name_text = match.group(0)
else:
    org_name_text = ""
# Find the div with class "ocr_carea" and title attribute containing "bbox 489 1387 737 1420"
div_date_tag = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 489 1387 737 1420" in value)

# Extract text from the second div with the specified format (\d+,\s\w+\s\d+)
if div_date_tag:
    date_text = ""
    text = div_date_tag.get_text(strip=True)
    match = re.search(r"\b\d{1,2},\s\w+\s\d{4}\b", text)
    if match:
        date_text = match.group(0)
else:
    date_text = ""
div_p1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 127 260 994 325" in value)

# Extract text from the first div
if div_p1:
    p1 = div_p1.get_text(strip=True)
else:
    p1 = ""
div_h1 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 126 378 674 425" in value)

# Extract text from the first div
if div_h1:
    h1 = div_h1.get_text(strip=True)
else:
    h1 = ""

# Define a list of target IDs
target_ids = ["bbox 127 413 992 465", "bbox 124 502 877 668"]

# Initialize a variable to store the extracted text
p2= ""

# Loop over each target ID
for target_id in target_ids:
    # Find the div element with the specified class and title
    div = soup.find("div", class_="ocr_carea", title=lambda value: value and target_id in value)

    # Extract text from the div if found
    if div:
        p2 += div.get_text(strip=True) + "\n"

div_h2 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 123 709 987 808" in value)

# Extract text from the first div
if div_h2:
    h2 = div_h2.get_text(strip=True)
else:
    h2 = ""

# Define a list of target IDs
target_ids = ["bbox 118 791 987 853", "bbox 122 826 986 879","bbox 123 859 512 894"]

# Initialize a variable to store the extracted text
p3= ""

# Loop over each target ID
for target_id in target_ids:
    # Find the div element with the specified class and title
    div = soup.find("div", class_="ocr_carea", title=lambda value: value and target_id in value)

    # Extract text from the div if found
    if div:
        p3 += div.get_text(strip=True) + "\n"

div_h3 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 120 911 277 949" in value)

# Extract text from the first div
if div_h3:
    h3 = div_h3.get_text(strip=True)
else:
    h3 = ""

# Define a list of target IDs
target_ids = ["bbox 120 960 985 1023", "bbox 115 1031 983 1088","bbox 111 1099 981 1273","bbox 109 1279 980 1448"]

# Initialize a variable to store the extracted text
p4= ""

# Loop over each target ID
for target_id in target_ids:
    # Find the div element with the specified class and title
    div = soup.find("div", class_="ocr_carea", title=lambda value: value and target_id in value)

    # Extract text from the div if found
    if div:
        p4 += div.get_text(strip=True) + "\n"


div_p5 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 168 110 759 323" in value)

# Extract text from the first div
if div_p5:
    p5 = div_p5.get_text(strip=True)
else:
    p5 = ""

div_h4 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 170 355 332 388" in value)

# Extract text from the first div
if div_h4:
    h4 = div_h4.get_text(strip=True)
else:
    h4 = ""


div_p6 = soup.find("div", class_="ocr_carea", title=lambda value: value and "bbox 172 405 474 439" in value)

# Extract text from the first div
if div_p6:
    p6 = div_p6.get_text(strip=True)
else:
    p6 = ""



# Create TEI XML document
tei_xml = f"""
<TEI>
  <teiHeader>
    <fileDesc>
    <titleStmt>
        <title>{title}</title>
        </titleStmt>
        <orgName>{org_name_text}</orgName>
      <publicationStmt>
        <publPlace>{place_text}</publPlace>
        <date>{date_text}</date>
      </publicationStmt>
    </fileDesc>
  </teiHeader>
  <text>
    <body>
      <div>
        <p>{p1}</p>
         <head>{h1}</head>
        <p>{p2}</p>
         <head>{h2}</head>
         <p>{p3}</p>
         <head>{h3}</head>
          <p>{p4}</p>
      </div>
      <div>
        <p>{p5}</p>
        <head>{h4}</head>
        <p>{p6}</p>
      </div>
    </body>
  </text>
</TEI>
"""

# Write TEI XML document to file
with open("output_141813.tei.xml", "w", encoding="utf-8") as file:
    file.write(tei_xml)
```
